```python
import pandas as pd


data = pd.DataFrame(index=pd.date_range("2021-01-01", periods=49, freq="MS"))


def calculate_users_monthly(begining_amount, churn_rates_dict):
    
    churn_dic = churn_dict(churn_rates_dict)
    for i in range(1, 49):
        begining_amount.append(begining_amount[i-1] * (1-churn_dic[i])**(1/12))
    return begining_amount


def calculate_users_annual(begining_amount, churn_rates_dict):
    
    churn_dic = churn_dict(churn_rates_dict)
    for i in range(1, 49):
        begining_amount.append(begining_amount[i-1] * (1-churn_dic[i])**(1/12))
    return begining_amount
    

    
def churn_dict(churn_dict):
    new_dict = {}
    keys_list = churn_dict.keys()
    for key in keys_list:
        values = list(map(int, key.split('-')))
        new_keys = list(range(values[0], values[1]+1))
        for new_key in new_keys:
            new_dict[new_key] = churn_dict[key]
            
    return new_dict



def calculate_users_annual(begining_amount, churn_rates_dict):
    
    churn_dic = churn_dict(churn_rates_dict)
    for i in range(12, 48, 12):
        value = begining_amount[i-1] * (1-churn_dic[i])
        for j in range(12):
            begining_amount.append(value)
     
    begining_amount.append(begining_amount[-1] * (1-churn_dic[list(churn_dic)[-1]]))
    return begining_amount

def calculate_revenue_monthly(data, string, price_intro, price_regular):
    revenue = [data[string][i]*price_intro for i in range(6)]
    for i in range(6,49):
        revenue.append(data[string][i]*price_regular)
        
    return revenue

def calculate_revenue_annual(data, string, price_intro, price_regular):
    revenue = [data[string][0]*price_intro]
    for i in range(1,49):
        if i%12==0:
            revenue.append(data[string][i]*price_regular)
        else:
            revenue.append(0)
        
    return revenue
    


data['Monthly_Basic'] = calculate_users_monthly([1800], {'1-6': 0.4, '7-12': 0.3, '13-24': 0.25, '25-36': 0.2, '37-48': 0.175})
data['Monthly_Pro'] = calculate_users_monthly([3000], {'1-6': 0.35, '7-12': 0.275, '13-24': 0.2, '25-36': 0.2, '37-48': 0.15})
data['Monthly_Premium'] = calculate_users_monthly([1200], {'1-6': 0.3, '7-12': 0.225, '13-24': 0.2, '25-36': 0.185, '37-48': 0.14})


data['Annual_Basic'] = calculate_users_annual(12*[1200], {'1-12': 0.4, '13-24': 0.25, '25-36': 0.2, '37-48': 0.175})
data['Annual_Pro'] = calculate_users_annual(12*[2000], {'1-12': 0.35, '13-24': 0.2, '25-36': 0.2, '37-48': 0.15})
data['Annual_Premium'] = calculate_users_annual(12*[800], {'1-12': 0.3, '13-24': 0.2, '25-36': 0.185, '37-48': 0.14})


upfront_costs = 15*(1200+1800) + 25*(3000+2000) + 40*(1200+800)


data['Monthly_Basic_Costs'] = 1.5 * data['Monthly_Basic']
data['Monthly_Pro_Costs'] = 2 * data['Monthly_Pro']
data['Monthly_Premium_Costs'] = 2.5 * data['Monthly_Premium']

data['Annual_Basic_Costs'] = 1.5 * data['Annual_Basic']
data['Annual_Pro_Costs'] = 2 * data['Annual_Pro']
data['Annual_Premium_Costs'] = 2.5 * data['Annual_Premium']

data['Revenue_Basic_Monthly'] = calculate_revenue_monthly(data, 'Monthly_Basic', 5, 9)
data['Revenue_Pro_Monthly'] = calculate_revenue_monthly(data, 'Monthly_Pro', 9, 19)
data['Revenue_Premium_Monthly'] = calculate_revenue_monthly(data, 'Monthly_Premium', 19, 39)


data['Revenue_Basic_Annual'] = calculate_revenue_annual(data, 'Annual_Basic', 49, 99)
data['Revenue_Pro_Annual'] = calculate_revenue_annual(data, 'Annual_Pro', 99, 199)
data['Revenue_Premium_Annual'] = calculate_revenue_annual(data, 'Annual_Premium', 199, 399)


print(upfront_costs)
data
```

```python
costs = data[[x for x in data.columns if 'Costs' in x]]

costs_total = costs[:-1].sum()
costs_total.sum()
```

```python
new = data[:-1]

revenue = new[[x for x in new.columns if 'Revenue' in x]].sum()


costs = new[[x for x in new.columns if 'Costs' in x]].sum()


cashflow = revenue.sum() - costs.sum() - upfront_costs

cashflow.sum()
```

```python
revenue = new[[x for x in new.columns if 'Revenue' in x]]


costs = new[[x for x in new.columns if 'Costs' in x]]

discount_factors = [1/((1+0.12)**(i/12)) for i in range(48)]


revenue = revenue.sum(axis=1)
revenue[0] = revenue[0] - upfront_costs
cashflow = revenue - costs.sum(axis=1)
(cashflow*discount_factors).sum()
```
